```python
#Defining the Base Path
dataset_path = '/content/drive/MyDrive/G05 Miniproject/Datasets-20250212T185648Z-001/Datasets/Updated Dataset'
```

```python
#Listing Files in the Dataset Directory
sites = os.listdir(dataset_path)
print('Total number of sites:', len(sites))
```

```python
df = []
for site in sites:
    file_path = os.path.join(dataset_path, site)
    try:
        # Read the CSV while skipping rows
        data = pd.read_csv(file_path, skiprows=10)

        # Check if the file is empty after skipping rows
        if data.empty:
            print(f"Skipped empty file: {site}")
            continue

        # Append row count if data exists
        row_count = len(data)
        df.append(row_count)

    except pd.errors.EmptyDataError:
        print(f"Empty file encountered: {site}")
        continue  # Skip the file and move to the next

df
```

```python
a = np.array(df)
a.min(), a.mean(), a.std(),
```

```python

df = None  # Initialize as None to start the concatenation
for site in sites:
    file_path = os.path.join(dataset_path, site)

    try:
        # Read the CSV file while skipping 10 rows
        temp_df = pd.read_csv(file_path, skiprows=10)

        # Skip if the DataFrame is empty after reading
        if temp_df.empty:
            print(f"Skipped empty file or file with no data after skipping rows: {site}")
            continue

        # Concatenate if df is not None, otherwise initialize df
        if df is not None:
            df = pd.concat((df, temp_df), ignore_index=True)
        else:
            df = temp_df

    except pd.errors.EmptyDataError:
        print(f"EmptyDataError: No data found in file: {site}")
        continue  # Skip files causing the error

df
```

```python
print('Total {:d} records found.'.format(len(df)))
```

```python
# Drop any example with missing target value
df.dropna(how='all', axis='index', subset=['GHI (Wh/m2)'], inplace=True)
df
```

```python
# Checking missing values
for col in df.columns:
    if sum(pd.isnull(df[col]).values) > 0:
        print('{0:s} has {1:d} missing values.'.format(col, sum(pd.isnull(df[col]).values)))
```

```python
# Drop records with missing values
df_dropped = df.dropna(how='any', axis='index')
df_dropped
```
